```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from openai import OpenAI
openai_client = OpenAI()
```

```python
def llm(prompt):
    response = openai_client.responses.create(
        model = 'gpt-5.4-mini',
        input = prompt
    )
    return response.output_text
```

```python
llm('hey whats up')
```
